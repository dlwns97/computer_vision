```python
import tensorflow as tf

x_data = [[1,2],
          [2,3],
          [3,1],
          [4,3],
          [5,3],
          [6,2]]
y_data = [[0],
          [0],
          [0],
          [1],
          [1],
          [1]]

tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=1,input_dim=2))
# use sigmoid function as Activation for 0~1 problem
tf.model.add(tf.keras.layers.Activation('sigmoid'))

"""
cost function을 mse로 사용하는 것보다
binary_crossentropy를 사용하는 것이 훨씬 좋은 결과를 낼 수 있다.

성능 척도 metrics의 특성으로 'accuracy'를 준다.
"""
tf.model.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.SGD(lr=0.01), metrics=['accuracy'])
tf.model.summary()

history = tf.model.fit(x_data, y_data, epochs=5000)

# Accuracy report

print("Accuracy: ", history.history['accuracy'][-1])
```

```python
import numpy as np

xy = np.loadtxt('./drive/MyDrive/machine learning/data/data-03-diabetes.csv', delimiter=',', dtype=np.float32)
x = xy[:,0:-1]
y = xy[:,[-1]]

"""
print(x.shape, y.shape) 
데이터를 받아오면 데이터의 형태를 확인
"""

tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=x.shape[1], activation='sigmoid'))
tf.model.compile(loss='binary_crossentropy', optimizer = tf.keras.optimizers.SGD(lr=0.01), metrics=['accuracy'])
tf.model.summary()

history = tf.model.fit(x, y, epochs=500)

# 정확도
print("Accuracy: ".format(history.history['accuracy'][-1]))

# 테스트 케이스 예측
y_predict = tf.model.predict([[0.17641,0.155779, 0, 0, 0, 0.052161, -0.952178, -0.733333]])
print("Predict: {0}".format(y_predict))

# evaluating model
"""
evaluate = tf.model.evaluate(x, y)
print("loss: {0}, accuracy: {1}".format(evaluate[0], evaluate[1]))
"""
```
