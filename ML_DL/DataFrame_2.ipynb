```python
# 새로운 조건에 맞는 변수 추가
abalone_df['length_bool'] = np.where(abalone_df['length']>abalone_df['length'].median(),
                                     'length_long', # True
                                     'length_short') # False
```

```python
# NaN(결측치)를 찾아서 다른 값으로 변경
# 기존 데이터에는 결측치가 존재 X
abalone_df.isnull().sum().sum()
```

```python
# 가상으로 결측치 만들기
nan_abalone_df = abalone_df.copy()
nan_abalone_df.loc[2, 'length'] = np.nan
nan_abalone_df.loc[3,'rings'] = np.nan
nan_abalone_df.loc[0,'sex'] = np.nan
nan_abalone_df.isnull().sum().sum()
```

```python
# 결측치를 특정 값으로 채우기
# 실수가 정수를 이김, 0으로 채우면 0.000
zero_abalone_df = nan_abalone_df.fillna(0)
zero_abalone_df
```

```python
# 결측치를 결측치가 속한 컬럼의 평균값으로 대체하기
nan_abalone_df.mean()
```

```python
# 알아서 해당 column의 평균 값으로 들어감
# 연속형 변수만 이 방식으로 채울 수 있음
nan_abalone_df = nan_abalone_df.fillna(nan_abalone_df.mean())
# categorical한 값은 평균으로 채울 수 가 없음!
sex_mode = nan_abalone_df['sex'].value_counts().sort_values(ascending=False).index[0]
nan_abalone_df['sex'] = nan_abalone_df['sex'].fillna(sex_mode)
nan_abalone_df
```

```python
# 데이터 프레임의 꽃
# apply 함수
# 본인이 원하는 행과 열에 연산 혹은 function을 적용할 수 있다.
# 열 기준 axis = 0
# 행 기준 axis = 1
abalone_df[['diameter']].apply(np.average, axis = 0)
```

```python
# 행 기준 집계
abalone_df[['diameter', 'whole_weight']].apply(np.average, axis=1)
```

```python
# 사용자 함수를 통한 집계 ★★★★★
import math
def avg_ceil(x, y, z):
   return math.ceil((x+y+z))
abalone_df[['diameter','height','whole_weight']].apply(lambda x: avg_ceil(x[0],x[1],x[2]), axis=1)
```

```python
# 사용자 정의 함수 사용
# ['diameter','height', 'whole_weight'] 변수 사용
# 세 변수의 합이 1이 넘으면 True, 아니면 False 출력 후 answer 변수에 저장
# abalone_df에 answer 열을 추가하고 입력
abalone_sum = abalone_df[['diameter','height','whole_weight']].apply(lambda x: x[0]+x[1]+x[2], axis=1)>1
abalone_sum.name = 'answer'
pd.concat([abalone_df, abalone_sum], axis=1)
```
